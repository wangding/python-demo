例：利用无穷级数展开式求$\pi$的近似值

$\frac{\pi}{4} = 1 - \frac{1}{3} + \frac{1}{5} - \frac{1}{7} + \cdots + (-1)^{n+1}\frac{1}{2n-1}$


这个公式就是著名的**莱布尼茨公式**（Leibniz formula for π），通过交错级数来逼近$\frac{\pi}{4}$的值。当n趋向无穷大时，级数的和收敛于$\frac{\pi}{4}$。

```python
# 向量编程风格，避免使用 for 循环
import numpy as np

n = 1000
x = np.arange(1, 2*n, 2)
y = (-1) ** np.arange(2, n+2) / x
np.sum(y) * 4
```
