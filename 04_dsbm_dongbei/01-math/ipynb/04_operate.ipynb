# 标量、向量、矩阵的基本运算

```python
import numpy as np
```

```python
# 标量乘向量
a = np.array([4, 10, 6])
a, 2 * a
```

```python
# 标量乘矩阵
b = np.array([[4, 10, 6], [-2, 4, 0], [0, 4, 5]])
b, 2 * b
```

```python
# 向量加向量
c = np.array([4, 10, 6])
d = np.array([1, 2, 3])
c, d, c + d
```

```python
# 矩阵加矩阵
e = np.array([[4, 10, 6], [-2, 4, 0], [0, 4, 5]])
f = np.array([[1, 2, 3], [3, 4, 5], [7, 8, 9]])
e, f, e + f
```

```python
# 矩阵 + 行向量，广播，行向量加到矩阵的每一行
e, d, e + d
```

# 向量的乘法或内积

```python
g = np.array([1, 0, 3])
h = np.array([1, 2, 3])
g, h
```

```python
# 向量内积
np.dot(g, h)
```

```python
# 向量叉积
np.cross(g, h)
```

```python
# 验证叉积与原向量垂直
np.dot(np.cross(g, h), g)
np.dot(np.cross(g, h), h)
```

```python
# 向量点乘，对应位相乘
g * h
```

# 矩阵乘法

```python
j = np.array([[1, 0, 3], [0, 2, 4]])
k = np.array([[1, 1], [2, 0], [3, 2]])
j, k
```

```python
# 矩阵相乘
np.dot(j, k)
```

```python
# 矩阵乘法应用

# 产品a 和 产品b 的三项成本构成
l = np.array([[1, 1], [2, 3], [3, 5]])
# 产品a 和 产品b 四个季度的产量
m = np.array([[100, 150, 180, 120], [300, 200, 240, 180]])

# 求两个产品在每个季度的总成本
quarterly_costs = np.dot(l, m)
quarterly_costs
```

# 矩阵除法和逆矩阵

```python
n = np.array([[2, 3], [4, 5]])
o = np.linalg.inv(n)
n, o
```

```python
# 验证逆矩阵
np.dot(n, o)
```
