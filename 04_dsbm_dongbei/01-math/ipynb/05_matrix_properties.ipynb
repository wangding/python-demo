# 矩阵和向量属性

```python
import numpy as np
```

## 函数

函数是映射 $f: A \to B$

```python
# 函数的参数可以是标量
x = 4
y = np.sqrt(x)
z = np.sin(y)
y, z
```

```python
# 函数的参数可以是向量
a = np.array([4, 5, 6])
b = np.array([1, 2, 3])
np.dot(a, b)  # 向量点积
```

```python
# 函数的参数可以是矩阵
d = np.array([[1, 2, 3], [2, 4, 6], [3, 6, 9]])
np.diag(d)  # 主对角线
```

## 向量范数

```python
v = np.array([2, 3, -5, -7])
```

1范数：向量各元素绝对值之和

$\| \mathbf{x} \|_1 = \sum_{i=1}^{n} \lvert x_i \rvert$

```python
np.linalg.norm(v, 1)
```

2范数：欧几里得范数（向量模长）

$\| \mathbf{x} \|_2 = \sqrt{ \sum_{i=1}^{n} \lvert x_i \rvert^2 }$

```python
np.linalg.norm(v, 2)
```

p范数

$\| \mathbf{x} \|_p = \left( \sum_{i=1}^{n} \lvert x_i \rvert^p \right)^{1/p}$

```python
p = 3
np.linalg.norm(v, p)
```

正无穷范数：向量元素绝对值的最大值

$\| \mathbf{x} \|_\infty = \max_{1 \le i \le n} \lvert x_i \rvert$

```python
np.linalg.norm(v, np.inf)
```

负无穷范数：向量元素绝对值的最小值

$\| \mathbf{x} \|_{-\infty} = \min_{1 \le i \le n} \lvert x_i \rvert$

```python
np.linalg.norm(v, -np.inf)
```

## 矩阵范数

```python
e = np.array([[-1, 0, 5], [2, 4, 6], [0, 4, 5]])
np.linalg.norm(e, 1)
np.linalg.norm(e, 2)
np.linalg.norm(e, np.inf)
```

## 特征值和特征向量

```python
# 特征值和特征向量
f = np.array([[2, 1, 0], [1, 3, 1], [0, 1, 4]])
vals, vecs = np.linalg.eig(f)
```
