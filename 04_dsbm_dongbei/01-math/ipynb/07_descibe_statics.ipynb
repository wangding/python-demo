# 描述性统计

```python
# 导入数据科学三大包
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
scores = np.array([
  76, 81, 85, 79, 83, 82, 91, 88, 87, 86,
  70, 82, 84, 77, 83, 85, 76, 74, 88, 80,
  82, 76, 68, 77, 80, 78, 77, 73, 81, 76,
  85, 81, 84, 85, 74, 84, 70, 76, 78, 80,
  86, 75, 94, 79, 84, 78, 72, 86, 74, 68
])

# 数据科学基础课程的成绩
scores
```

```python
# 算数平均数
np.mean(scores)
```

```python
# 几何平均数
stats.gmean(scores)
```

```python
# 调和平均数
stats.hmean(scores)
```

```python
# 去尾平均
stats.tmean(scores, (70, 90), (True, False))
```

```python
# 中位数
np.median(scores)
# np.quantile(scores, 0.5)
```

```python
# 众数
stats.mode(scores)
```

```python
# 极值（最小值和最大值，0% 分位数 和 100% 分位数）
np.quantile(scores, [0, 1])
#np.amin(scores), np.amax(scores)
```

```python
# 极差 peak to peak
np.ptp(scores)
```

```python
# 四分位距离 inter quartile range
stats.iqr(scores)
#np.quantile(scores, 0.75) - np.quantile(scores, 0.25)
```

```python
# 总体方差 variance
np.var(scores)
```

```python
# 样本方差 delta degree of freedom = 1
np.var(scores, ddof=1)
```

```python
# 总体标准差 standard deviation
np.std(scores)
```

```python
# 样本标准差
np.std(scores, ddof=1)
```

```python
# 变异系数 coefficient of variation
stats.variation(scores)
#np.std(scores) / np.mean(scores)
```

```python
# 箱线图 box plot
plt.boxplot(scores, showmeans=True, notch=True)
plt.ylim(0, 100)
plt.show()
```

```python
# 直方图 histogram
plt.hist(scores, bins=np.arange(65, 101, 5))
plt.show()
```

```python
import seaborn as sns
sns.histplot(scores, bins=np.arange(65, 101, 5), kde=True)
plt.show()
```

```python
# 偏度系数 skewness
stats.skew(scores)
```

```python
# 峰度系数 kurtosis
stats.kurtosis(scores)
```

```python
scores = np.random.randint(60, 101, (10, 3))

# 10 位学生，三门课程的成绩表
scores
```

```python
# 按列计算均值
np.mean(scores, axis=0)
```

```python
# 按行计算均值
np.mean(scores, axis=1).round(1)
```

```python
# 默认按列计算变异系数，相当于 axis=0, 还可以 axis=1, axis=None
stats.variation(scores)
```

```python
# 按列计算方差
np.var(scores, axis=0)
```

```python
# 描述性统计汇总
stats.describe(scores)
```
