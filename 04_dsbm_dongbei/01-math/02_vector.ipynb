# 向量

**向量**是一组有序排列的数。通过次序索引，可以得到每个单独的数（元素）。  

行向量：
$
X= \begin{bmatrix}
x_1 & x_2 & \cdots & x_n
\end{bmatrix}
$

列向量：
$
X= \begin{bmatrix}
x_1 \\
x_2 \\
\vdots \\
x_n
\end{bmatrix}
$

向量中的元素可以代表：
- 某个事物的若干种状态
- 某个对象的若干属性
- 某个多项式的若干个系数$

```python
import numpy as np
```

```python
# 行向量
np.array([1, 2, 3, 4, 5])
```

```python
# 列向量
b = np.array([[1], [2], [3], [4], [5]])
b
```

```python
# 向量转置
b.T
```

```python
# 冒号生成向量
np.arange(1, 6)
```

```python
# 冒号指定步长
np.arange(1, 11, 2)
```

```python
# 函数生成向量，默认数量
np.linspace(1, 10)
```

```python
# 指定数量
np.linspace(1, 10, 5)
```

```python
# 对数等分
np.logspace(1, 5, 3)
```
