# 函数的属性

```python
# 导入三大数据科学包
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, limit, diff, atan, latex, derive_by_array
import sympy as sp

from IPython.display import Math, display
```

## 1. 奇偶性

### 偶函数

偶函数满足：$f(-x) = f(x)$，轴对称  
例如：$y = x^2$

### 奇函数

奇函数满足：$f(-x) = -f(x)$，原点对称  
例如：$y = x^3$

```python
# 绘制 x^2 偶函数和 x^3 奇函数图像
x = np.linspace(-5, 5, 100)
y1 = x**2  # 偶函数
y2 = x**3  # 奇函数

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# 绘制y = x^2
ax1.plot(x, y1, 'b-', linewidth=2)
ax1.grid(True)
ax1.set_title('y = x^2')

# 绘制y = x^3
ax2.plot(x, y2, 'r--', linewidth=2)
ax2.grid(True)
ax2.set_title('y = x^3')

plt.tight_layout()
plt.show()
```

## 2. 周期性

周期函数满足：$f(x+T) = f(x)$  
例如：$y = sin(x)$

```python
# 绘制 y = sin(x) 周期函数图像
x = np.linspace(-10, 10, 200)
y = np.sin(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', linewidth=2)
plt.grid(True)
plt.title('y = sin(x)')
plt.axis('tight')

# 标记周期
for i in range(-1, 2):
    plt.axvline(x=i*2*np.pi, color='r', linestyle='--', alpha=0.3)
    plt.text(i*2*np.pi, 1.1, f'{i}×2π', ha='center')

plt.ylim(-1.5, 1.5)
plt.show()
```

## 3. 单调性

### 单调增函数

对于定义域内任意 $x_1 < x_2$，恒有 $f(x_1) < f(x_2)$  
例如：$y=log(x)$

### 单调减函数

对于定义域内任意 $x_1 < x_2$，恒有 $f(x_1) > f(x_2)$  
例如：$y=-log(x)$

```python
# 绘制 log(x) 单调增函数和 -log(x) 单调减函数图像
x = np.linspace(0.01, 3, 100)
y1 =  np.log(x)  # 单调增函数
y2 = -np.log(x)  # 单调减函数

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# 绘制 y = log(x)
ax1.plot(x, y1, 'b-', linewidth=2)
ax1.grid(True)
ax1.set_title('y = log(x)')

# 绘制 y = -log(x)
ax2.plot(x, y2, 'r--', linewidth=2)
ax2.grid(True)
ax2.set_title('y = -log(x)')

plt.tight_layout()
plt.show()
```

```python
# 单调性通过导数判断
x_sym = symbols('x')

# f(x) = x^3 的导数
f1 = x_sym**3
f1_prime = diff(f1, x_sym)

display(Math(f"f(x) = {latex(f1)}"))
display(Math(f"f'(x) = {latex(f1_prime)}"))
display(Math(f"由于 f'(x) = {latex(f1_prime)} ≥ 0  对所有实数 x 成立，因此 f(x) = {latex(f1)} 是单调增函数"))

# f(x) = -x 的导数
f2 = -x_sym
f2_prime = diff(f2, x_sym)

print()
display(Math(f"f(x) = {latex(f2)}"))
display(Math(f"f'(x) = {latex(f2_prime)}"))
display(Math(f"由于 f'(x) = {latex(f2_prime)} < 0 对所有实数x成立，因此 f(x) = {latex(f2)} 是单调减函数"))
```

## 4. 连续性

函数在点 $x_0$ 处连续，当且仅当：

$\lim_{x \to x_0^-} f(x) = \lim_{x \to x_0^+} f(x) = f(x_0)$

或者用 $\epsilon$-$\delta$ 语言表示：对于任意 $\epsilon > 0$，存在 $\delta > 0$，

当 $|x - x_0| < \delta$ 时，恒有 $|f(x) - f(x_0)| < \epsilon$

```python
# 连续性验证示例
x_sym = symbols('x')

# 检查 f(x) = x^2 在 x=0 处的连续性
f = x_sym**2
limit_left = limit(f, x_sym, 0, dir='-')  # 左极限
limit_right = limit(f, x_sym, 0, dir='+')  # 右极限
f_at_0 = f.subs(x_sym, 0)  # 函数在 x=0 处的值

display(Math(f"f(x) = {latex(f)}"))
display(Math(f"左极限: lim_{{x→0⁻}} f(x) = {limit_left}"))
display(Math(f"右极限: lim_{{x→0⁺}} f(x) = {limit_right}"))
display(Math(f"函数值: f(0) = {f_at_0}"))
display(Math(f"连续性: {limit_left == limit_right == f_at_0}"))
```

## 5. 极限

如果当 $x$ 无限接近 $a$ 时，函数 $f(x)$ 无限接近常数 $L$，则称：

$\lim_{x \to a} f(x) = L$

对于无穷远点的极限：

$\lim_{x \to \infty} f(x) = L$ 表示当 $x$ 趋向正无穷时，$f(x)$ 趋向于 $L$

```python
# 计算极限示例
x_sym = symbols('x')
y = atan(x_sym)

# 计算 x 趋向负无穷时的极限
limit_neg_inf = limit(y, x_sym, '-oo')

# 计算 x 趋向正无穷时的极限
limit_pos_inf = limit(y, x_sym, 'oo')

display(Math(f"f(x) = {y}"))
display(Math(f"lim_{{x→-∞}} f(x) = {latex(limit_neg_inf)}"))
display(Math(f"lim_{{x→+∞}} f(x) = {latex(limit_pos_inf)}"))
```

```python
# 绘制 arctan(x) 函数及其渐近线
plt.rcParams['font.family'] = ['SimHei', 'sans-serif']

x = np.linspace(-20, 20, 200)
y = np.arctan(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', linewidth=2)
plt.axhline(y=-np.pi/2, color='r', linestyle='--', alpha=0.5, label=f'y = -π/2 ≈ {-np.pi/2:.4f}')
plt.axhline(y=np.pi/2, color='r', linestyle='--', alpha=0.5, label=f'y = π/2 ≈ {np.pi/2:.4f}')
plt.axhline(y=0, color='k', linestyle='-', linewidth=0.5)
plt.axvline(x=0, color='k', linestyle='-', linewidth=0.5)
plt.grid(True)
plt.title('y = arctan(x) 及其渐近线')
plt.legend()
plt.ylim(-2, 2)
plt.show()
```

## 6. 导数

函数 $f(x)$ 在点 $x_0$ 处的导数定义为：

$f'(x_0) = \lim_{\Delta x \to 0} \frac{f(x_0 + \Delta x) - f(x_0)}{\Delta x}$

导数的几何意义是函数图像在该点处切线的斜率，物理意义是函数在该点的变化率

```python
# 计算导数示例
x_sym = symbols('x')
y = atan(x_sym)

# 计算导数
derivative = diff(y, x_sym)

display(Math(f"f(x) = {latex(y)}"))
display(Math(f"f'(x) = {latex(derivative)}"))
```

```python
# 绘制 arctan(x) 及其导数
plt.rcParams['font.family'] = ['SimHei', 'sans-serif']

x = np.linspace(-5, 5, 100)
y = np.arctan(x)
y_prime = 1 / (1 + x**2)  # arctan(x) 的导数

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', linewidth=2, label='y = arctan(x)')
plt.plot(x, y_prime, 'r--', linewidth=2, label="y' = 1/(1+x²)")
plt.axhline(y=0, color='k', linestyle='-', linewidth=0.5)
plt.axvline(x=0, color='k', linestyle='-', linewidth=0.5)
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.title('y = arctan(x) 和 y 的导数')
plt.legend()
plt.show()
```

## 7. 偏导数

对于二元函数 $f(x, y)$，关于 $x$ 的偏导数定义为：

$\frac{\partial f}{\partial x} = \lim_{\Delta x \to 0} \frac{f(x + \Delta x, y) - f(x, y)}{\Delta x}$

关于 $y$ 的偏导数定义为：

$\frac{\partial f}{\partial y} = \lim_{\Delta y \to 0} \frac{f(x, y + \Delta y) - f(x, y)}{\Delta y}$

```python
# 计算偏导数示例
x_sym, y_sym = symbols('x y')
f = x_sym**y_sym - y_sym**x_sym

# 计算对 x 的偏导数
df_dx = diff(f, x_sym)

# 计算对 y 的偏导数
df_dy = diff(f, y_sym)

display(Math(f"f(x, y) = {latex(f)}"))
display(Math(rf"\frac{{\partial f}}{{\partial x}} = {latex(df_dx)}"))
display(Math(rf"\frac{{\partial f}}{{\partial y}} = {latex(df_dy)}"))
```

## 8. 梯度

对于多元函数 $f(x_1, x_2, ..., x_n)$，梯度是一个向量，定义为：

$\nabla f = \left( \frac{\partial f}{\partial x_1}, \frac{\partial f}{\partial x_2}, ..., \frac{\partial f}{\partial x_n} \right)$

梯度的方向是函数增长最快的方向，梯度的模长是函数在该方向上的最大变化率

```python
# 计算梯度示例
x_sym, y_sym = symbols('x y')
f = x_sym**2 * y_sym**3

# 计算梯度
grad_f = derive_by_array(f, (x_sym, y_sym))

display(Math(f"f(x, y) = {latex(f)}"))
display(Math(f"\\nabla f = {latex(grad_f)}"))
```

```python
# 绘制 f(x, y) = x²y³ 的梯度场
plt.rcParams['font.family'] = ['SimHei', 'sans-serif']

# 创建网格
x = np.linspace(-2, 2, 10)
y = np.linspace(-2, 2, 10)
X, Y = np.meshgrid(x, y)

# 计算函数值
Z = X**2 * Y**3

# 计算梯度分量
U = 2 * X * Y**3  # ∂f/∂x
V = 3 * X**2 * Y**2  # ∂f/∂y

# 绘制3D表面图
fig = plt.figure(figsize=(15, 6))

# 3D表面图
ax1 = fig.add_subplot(121, projection='3d')
surf = ax1.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_zlabel('z')
ax1.set_title('f(x, y) = x²y³ 表面图')
fig.colorbar(surf, ax=ax1, shrink=0.5, aspect=5)

# 梯度场图
ax2 = fig.add_subplot(122)
ax2.streamplot(X, Y, U, V, color='r', density=0.8, linewidth=1)
ax2.contour(X, Y, Z, 20, cmap='viridis')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('f(x, y) = x²y³ 梯度场和等高线')
ax2.grid(True)
ax2.set_aspect('equal')

plt.tight_layout()
plt.show()
```
