# 统计图表会说谎

案例：某公司去年下半年盈利情况。

| 月份   | 7月 | 8月 | 9月 | 10月 | 11月 | 12月 |
|--------|-----|-----|------|------|------|------|
| 例如（百万） | 2.0 | 2.1 | 2.2 | 2.1 | 2.3 | 2.4 |

**绘图方案一**：y 轴范围为 [0.0, 2.5]（**弱化波动，体现“稳定”**）

```python
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = ['SimHei', 'sans-serif']

# 1. 准备数据
months = ['7月', '8月', '9月', '10月', '11月', '12月']
profits = [2.0, 2.1, 2.2, 2.1, 2.3, 2.4]  # 单位：百万美元

# 2. 创建画布并绘制折线图
plt.figure(figsize=(6, 4))  # 设置画布大小（可选）
plt.plot(months, profits, marker='o', linestyle='-', color='blue')  # 绘制折线，添加标记点

# 3. 配置 y 轴范围（关键区别！）
plt.ylim(0.0, 2.5)  

# 4. 添加标签、标题等
plt.xlabel('月份')
plt.ylabel('利润（百万美元）')
plt.title('公司每月利润（y 轴范围：0.0 - 2.5）')
plt.grid(True)  # 添加网格线（可选，增强可读性）

# 5. 显示图形
plt.show()
```

**绘图方案二**：y 轴范围为 [2.0, 2.5]（**放大波动，体现“大幅增长”**）

```python
# 1. 数据与代码段 1 完全一致
# 2. 创建画布并绘制折线图
plt.figure(figsize=(6, 4))  
plt.plot(months, profits, marker='o', linestyle='-', color='blue')  

# 3. 配置 y 轴范围（关键区别！）
plt.ylim(2.0, 2.5)  

# 4. 添加标签、标题等
plt.xlabel('月份')
plt.ylabel('利润（百万美元）')
plt.title('公司每月利润（y 轴范围：2.0 - 2.5）')
plt.grid(True)  
```
