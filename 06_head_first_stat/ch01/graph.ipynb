# 饼图

“饼图”的作用是将数据划分为互有明显区别的几个组，或者叫做几个类。饼图为圆形，被分割为几个扇形块，每一块代表一个组（类）。
扇形块的大小表示这类数据占总体的比例。扇形块越大，该组（类）的相对频繁程度越大。一个特定组中的对象数目称为**频数**。
饼图将整个数据集划分为几个互不相干的组。这意味着，如果把每个扇形块的频数加起来，结果应为 100%。

**案例**：各种游戏软件销量数据如下：

| 游戏种类   | 销量（件） | 
|--------|-----|
| 体育 | 27500 |
| 策略 | 11500 |
| 动作 | 6000 |
| 射击 | 3500 |
| 其他 | 1500 |

```python
import matplotlib.pyplot as plt

# 游戏种类标签
labels = ['体育', '策略', '动作', '射击', '其他']  
# 各游戏销量（单位：件）
sizes = [27500, 11500, 6000, 3500, 1500]

# 设置画布大小（可选，按需调整）
plt.rcParams['font.family'] = ['SimHei', 'sans-serif']
plt.figure(figsize=(8, 6))  

# 绘制饼图
plt.pie(
    sizes,            # 销量数据
    labels=labels,    # 游戏种类标签
    autopct='%1.1f%%',# 显示百分比（保留1位小数）
    startangle=140    # 饼图起始角度（调整布局）
)  

# 添加图表标题
plt.title('各种游戏销量')  

# 确保饼图为正圆形（避免因画布拉伸变形）
plt.axis('equal')  

# 显示图形
plt.show()
```

# 垂直条形图

就像饼图一样，条形图能让你对相对大小进行比较，但条形图还有这样一个优点：更精确。对于各个类的大小大致相同的情况，条形图是理想的图形，你能更精确地指出哪个类的频数最高，也更容易发现细小的差别。
条形图中的每一个长方形代表一个特定类，长方形的长度代表某种数值。长方形越长，数值越大。所有长方形的宽度都相等，这样更容易进行比较。

**案例**：各款游戏软件在五个地区的销量数据如下：

| 地区   | 销量（件） | 
|--------|-----|
| A | 1000 |
| B | 5000 |
| C | 7500 |
| D | 8000 |
| E | 9500 |

```python
# 1. 准备数据
regions = ['A', 'B', 'C', 'D', 'E']
sales = [1000, 5000, 7500, 8000, 9500]

# 2. 绘制垂直条形图
bars = plt.bar(regions, sales, color='lightblue')

# 3. 遍历每个条形，添加销售数字标签
# for bar, sale in zip(bars, sales):
#     # 获取条形的水平中心（x 坐标）、高度（y 坐标）
#     bar_x = bar.get_x() + bar.get_width() / 2   # 条形水平中心
#     bar_height = bar.get_height()               # 条形高度
    
#     # 在条形顶部上方标注销售数字（y 坐标 = 条形高度 + 偏移量，避免与条形重叠）
#     plt.text(
#         x=bar_x, 
#         y=bar_height + 50,  # 偏移量设为 50（可根据需求调整）
#         s=f"{sale}",        # 标签文本（直接使用销售数字）
#         ha='center',        # 水平居中对齐
#         va='bottom'         # 垂直方向“底部”对齐（文本底部与 bar_height+10 对齐）
#     )

# 4. 添加标签与标题
plt.xlabel('地区')
plt.ylabel('销量（件）')
plt.title('各地区销量（件）')

# 5. 显示图表
plt.show()
```

# 水平条形图

**案例**：各类游戏玩家满意度数据如下：

| 游戏种类 | 满意率（%） | 
|--------|-----|
| 体育 | 99 |
| 策略 | 90 |
| 动作 | 85 |
| 射击 | 95 |
| 其他 | 80 |

```python
# 1. 准备数据
categories = ['体育', '策略', '动作', '射击', '其他']
satisfaction_rates = [99, 90, 85, 95, 80]

# 2. 绘制水平条形图
plt.barh(categories, satisfaction_rates, color='lightgreen')

# 3. 配置图表样式
plt.title('每种游戏的满意玩家(%)')
plt.xlabel('满意率（%）')
# plt.legend(['满意率（%）'])

# 4. 显示图表
plt.show()
```
